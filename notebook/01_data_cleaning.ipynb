```python
df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'].astype(str), format='%Y-%m')
```

```python
renewable_cols = ['Hydroelectric Power', 'Geothermal Energy', 'Solar Energy',
                  'Wind Energy', 'Wood Energy', 'Waste Energy', 'Fuel Ethanol, Excluding Denaturant']
```

```python
df['Total_Renewable'] = df[renewable_cols].sum(axis=1)
```

```python
sector_group = df.groupby(['Year', 'Sector'])[renewable_cols + ['Total_Renewable']].sum().reset_index()
```

```python
import os
import pandas as pd

project_root = os.path.join(os.getcwd(), "..")
data_dir = os.path.join(project_root, "data")

file_path = os.path.join(data_dir, "cleaned_us_renewable_energy.csv")
df_clean = pd.read_csv(file_path)
```
