```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

project_root = os.path.abspath(
    os.path.join(os.getcwd(), ".."))
file_path = os.path.join(project_root, "data", "cleaned_us_renewable_energy_no_2024.csv")

df_clean = pd.read_csv(file_path)
yearly_total = df_clean.groupby('Year')['Total_Renewable'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=yearly_total, x='Year', y='Total_Renewable', marker='o')
plt.title('US Total Renewable Energy Consumption (1973-2024')
plt.xlabel('Year')
plt.ylabel('Consumption (Trillion BTU)')
plt.grid(True)
plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

project_root = os.path.join(os.getcwd(), "..")
file_path = os.path.join(project_root, "data", "cleaned_us_renewable_energy_no_2024.csv")

df_clean = pd.read_csv(file_path)

df_clean.columns = df_clean.columns.str.strip()

renewable_cols = [
    'Hydroelectric Power',
    'Geothermal Energy',
    'Solar Energy',
    'Wind Energy',
    'Wood Energy',
    'Waste Energy',
    'Fuel Ethanol, Excluding Denaturant'
]

missing_cols = [col for col in renewable_cols if col not in df_clean.columns]
if missing_cols:
    print(f"Missing columns: {missing_cols}")
else:
    print("All columns available.")

yearly_total = df_clean.groupby('Year')[renewable_cols].sum().reset_index()

plt.figure(figsize=(14, 7))
for col in renewable_cols:
    plt.plot(yearly_total['Year'], yearly_total[col], label=col, marker='o', linewidth=2)

plt.title('US Total Renewable Energy Consumption by Source (19732024)')
plt.xlabel('Year')
plt.ylabel('Consumption (Trillion BTU)')
plt.legend()
plt.grid(True)
plt.show()
```
