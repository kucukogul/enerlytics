```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

project_root = os.path.abspath(
    os.path.join(os.getcwd(), ".."))
file_path = os.path.join(project_root, "data", "cleaned_us_renewable_energy.csv")

df_clean = pd.read_csv(file_path)
yearly_total = df_clean.groupby('Year')['Total_Renewable'].sum().reset_index()

plt.figure(figsize=(12,6))
sns.lineplot(data=yearly_total, x='Year', y='Total_Renewable', marker='o')
plt.title('US Total Renewable Energy Consumption (1973-2024')
plt.xlabel('Year')
plt.ylabel('Consumption (Trillion BTU?)')
plt.grid(True)
plt.show()
```
